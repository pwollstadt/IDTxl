#Using Cython interactively

```python
%load_ext cythonmagic
```

This loads the cythonmagic extensions.

This lets us use the cython magic together with the '--annotate' argument to get a profiled version of the code:

```python
%%cython --annotate

cdef int a = 0
for i in range(10):
    a += i
print(a)
```
